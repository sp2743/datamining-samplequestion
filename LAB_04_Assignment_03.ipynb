### 2. Use following techniques to handle missing data.  [40]
### a. mean imputation
### b. Class-wise mean imputation
### c. KNN imputation

### Compare the statistics after appling each imputations. Write your conclusion.

___________________________________________________________________________________________________________________________

```python
df1=df.copy()
df1.fillna(df1.mean(),inplace=True)
df1.head()
```

```python
df1.describe()
```

```python
df2=df.copy()
df2.fillna(df.groupby('target').transform('mean'), inplace=True)
df2.head()
```

```python
df2.describe()
```

```python
from sklearn.impute import KNNImputer
df3=df.copy()
def knnimputation(df3,k):
  impute=KNNImputer(n_neighbors=k)
  df3=impute.fit_transform(df3)
  return df3

df_imputed=knnimputation(df3,3)
df3=pd.DataFrame(df_imputed,columns=df.columns)
df3.head()
```

```python
df3.describe()
```

### 3. A list of DOBs given to you in a txt file.                                         [40]
### Convert all the DOBs in one common format:- YYYY-MM-DD

#### Example:-
    04-09-1978    1978-09-04
    Dec 11, 2001    2001-12-11
    1995/10/21    1995-10-21
    
### Display the output.

```python
from google.colab import files
uploaded=files.upload()
```

```python
lst=[]
with open("/content/DOBs (1).txt", "r") as file:
    content = file.read()

    # Split content by single quotes and extract dates
    parts = content.split("'")
    for i in range(1, len(parts), 2):
        lst.append(parts[i])

print(lst)
```

```python
date='07_21_1993'
cleaneddate=date.replace('_','-')
print(cleaneddate)
formatted_date=pd.to_datetime(cleaneddate)
print(formatted_date)
print(formatted_date.strftime('%Y-%m-%d'))
```

```python
cleaned_date=[]
for date in lst:
  cleaned_date.append(date.replace('_','-'))
cleaned_date
```

```python
dict={}
for date in cleaned_date:
  formatted_date=pd.to_datetime(date).strftime('%Y-%m-%d')
  dict[date]=formatted_date
dict
```

```python
df = pd.DataFrame(list(dict.items()), columns=['Original Date', 'Formatted Date'])
df.head()
```
